# Plotting with Julia using Gadfly and DataFrames

## General Workflow

### Step 1: Include Packages

```julia
using Gadfly, DataFrames
```

### Step 2: Get Data

```julia
using RDatasets
iris = dataset("datasets", "iris") # load the iris dataset
```

### Step 3: Plot, plot, plot!

```julia
plot(iris, x=:SepalLength, y=:SepalWidth, color=:Species, Geom.point)
```

## Getting Data

### From RDatasets

```julia
# there are lots of common benchmark-datasets available gathered in R-packages in julia you can use them with the following package
using RDatasets
RDatasets.datasets()
```

### From csv

```julia
# you can include data from .csv files with readtable
using DataFrames
pegel = readtable("./data/pegel.csv", separator=';')
```

### Filtering

```julia
# Eingangsdaten filtern
# mittelung
iris_mean = aggregate(iris, :Species, mean)
```
